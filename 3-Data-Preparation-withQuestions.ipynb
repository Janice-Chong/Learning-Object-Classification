### Ordinal encoding

#### Using self-defined mappings

#### 1. Demographics

```python
# get the demographics columns
# column 1 till column 3
col1_col3_df = df.iloc[:, :3]
col1_col3_df.head()
```

```python
# Define your custom mapping
custom_mapping = {
    "Male": 1,
    "Female": 2,
    "Postgraduate": 1,
    "Undergraduate": 2,
    "Certificate/Diploma": 3,
#     "Malaysia": 1,
#     "India": 0,
#     "Egypt": 0,
#     "Indonesia": 0,
#     "China": 0,
#     "Bangladesh": 0,
#     "Somalia": 0,
#     "Sudan": 0,
#     "Yemen": 0,
#     "Pakistan": 0,
#     "Less than RM 3000": 1,    #equalvalent to 'less than RM4849'
    "Less than RM 4,849": 1,   #equalvalent to 'less than RM4849'
#     "RM 3001 - 10 000": 2,     #equalvalent to 'RM 4,850 â€“ RM10,959'
    "RM 4,850 â€“ RM10,959": 2,  #equalvalent to 'RM 4,850 â€“ RM10,959'
#     "RM 10 001 - 25 000": 3,   #equalvalent to 'More than RM10,960'
    "More than RM10,960": 3,   #equalvalent to 'More than RM10,960'
}

# Perform ordinal encoding using the defined mapping
for column in col1_col3_df.columns:
    df[column] = df[column].map(custom_mapping)

# Display the encoded DataFrame
df.head()
```

#### 2. Learning objects

```python
# Get the learning objects column
learning_df = df.iloc[:, 5:18]
learning_df.head()
```

```python
custom_mapping = {
    "Not at All": 0,
    "Not Really": 0,
    "Undecided": 0,
    "Somewhat\t": 0,
    "Very Much": 1
}

# Perform ordinal encoding using the defined mapping
for column in learning_df.columns:
    df[column] = df[column].map(custom_mapping)

# Display the encoded DataFrame
df.head()
```

#### 3. Dominant VAK

```python
# Get the dominant VAK column for encoding
# column 48
domvak_df = df.iloc[:, 48:]
domvak_df.head()
```

```python
# Define your custom mapping
custom_mapping = {
    "Visual": 1,
    "Auditory": 2,
    "Kinesthetic": 3
}

# Perform ordinal encoding using the defined mapping
for column in domvak_df.columns:
    df[column] = df[column].map(custom_mapping)

# Display the encoded DataFrame
df.head()
```

### One-hot Encoding
Use one-hot encoding for the learning style questions columns which consist of 30 questions.

#### 1. VAK questions
- Extract columns
- Perform one-hot encoding

```python
# get the vak questions columns
# column 18 to column 47
qcolumns_df = df.iloc[:, 18:48]
qcolumns_df.head()
```

```python
# Encode categorical variables (one-hot encoding)
df = pd.get_dummies(df, columns=qcolumns_df.columns, prefix=qcolumns_df.columns)
df.head()
```

#### 2. Preferred learning mode

```python
prefered_columns = ['Preferred learning mode', 'Preferred Communication Platform']
# prefered_columns = ['Preferred Communication Platform']

# Encode categorical variables (one-hot encoding)
df = pd.get_dummies(df, columns=prefered_columns, prefix=prefered_columns)
df.head()
```
