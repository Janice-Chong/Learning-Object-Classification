### Data Granularity

#### Encoding the options into binary columns
For columns with multiple options, we will split the options into their respective columns. The value of the column would be 1 or 0 where:
- 1 = selected
- 0 = not selected

The column involved: Preferred learning mode  
Options: 
- Asynchronous Online Learning (On your own time)
- Face to Face
- Synchronous Online Learning (Real Time)

Then, new columns is append to the back of the table, and the original column is dropped.

```python
from sklearn.preprocessing import MultiLabelBinarizer
```

```python
# Splitting multiple options and creating lists of options
df['Preferred learning mode'] = df['Preferred learning mode'].str.split(', ')
# df['Preferred Social Media Platform '] = df['Preferred Social Media Platform '].str.split(', ')
# df['Preferred Communication Platform'] = df['Preferred Communication Platform'].str.split(', ')


# Initializing the MultiLabelBinarizer
mlb = MultiLabelBinarizer()


# Encoding the options into binary columns
encoded_data = pd.DataFrame(mlb.fit_transform(df['Preferred learning mode']), columns=mlb.classes_, index=df.index)
# Concatenating the original DataFrame with the encoded data
df_encoded = pd.concat([df, encoded_data], axis=1)

# encoded_data = pd.DataFrame(mlb.fit_transform(df['Preferred Social Media Platform ']), columns=mlb.classes_, index=df.index)
# df_encoded = pd.concat([df_encoded, encoded_data], axis=1)

# encoded_data = pd.DataFrame(mlb.fit_transform(df['Preferred Communication Platform']), columns=mlb.classes_, index=df.index)
# df_encoded = pd.concat([df_encoded, encoded_data], axis=1)


columns_to_remove = ["Preferred learning mode"]

df = df_encoded.drop(columns=columns_to_remove, axis=1)
df.head()
```

### Ordinal encoding

#### Using self-defined mappings

1. Extract the columns to be encoded: 
    - Gender 
    - Level of Study	
    - Country	
    - Household Income
    - Learning objects preference options
    - Dominant VAK

```python
# get the gender till learning objects perference columns
# column 1 till column 17
col1_col17_df = df.iloc[:, :17]
col1_col17_df.head()
```

```python
# get the dominant VAK column
# column 48 (last column)
vakcol_df = df.iloc[:, 47:48]
vakcol_df.head()
```

2. Join the 2 dataframes abaove contating the extracted columns that we want

```python
# Combine col1_col17_df and vakcol_df column-wise (axis=1)
columns_df = pd.concat([col1_col17_df, vakcol_df], axis=1)
columns_df.head()
```

### One-hot Encoding
Use one-hot encoding for the learning style questions columns which consist of 30 questions.

1. Extract the questions columns

```python
# get the learning objects perference columns
# column 18 to column 47
qcolumns_df = df.iloc[:, 17:47]
qcolumns_df.head()
```

2. Encode the data using get_dummies() function

```python
# Encode categorical variables (one-hot encoding)
features_encoded = pd.get_dummies(qcolumns_df)
features_encoded.head()
```

### Merging all encoded DataFrame into a single table

1. Extract demographics and learning objects columns

```python
lo_df = df.iloc[:, :17]
lo_df.head()
```

```python
# Combine lo_df and features_encoded column-wise (axis=1)
lo_ques_df = pd.concat([lo_df, features_encoded], axis=1)
lo_ques_df.head()
```

2. Extract the dominant VAK column and merge with the previously extracted columns

```python
back_df = df.iloc[:, 47:48]
back_df.head()
```

```python
# Combine lo_ques_df and the back_df column-wise (axis=1)
final_df = pd.concat([lo_ques_df, back_df], axis=1)
final_df.head()
```
