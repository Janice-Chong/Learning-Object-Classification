### Data Preparation
#### Extract columns for VAK Questionaires (30 Questions)

```python
# Extract columns 71 to 100
vak_df = df.iloc[:, 70:100]

# Display the selected columns
# print(vak_df)
vak_df.head()
```

#### Get survey responses 
To form MCQ options

```python

# Create a list to store lists of unique responses for each column
unique_responses_by_column = []

# Iterate through each column and extract unique responses
for column in vak_df.columns:
    unique_responses = vak_df[column].unique().tolist()
    unique_responses_by_column.append(unique_responses)

# Print the lists of unique responses grouped by column
for i, column in enumerate(vak_df.columns):
    print(f"Unique responses for column '{column}':")
    for j, response in enumerate(unique_responses_by_column[i]):
        print(f"{j + 1}. {response}")
    print()

# You can access unique responses for a specific column using unique_responses_by_column[column_index]
# For example, to access unique responses for the first column: unique_responses_by_column[0]
```

### Get User to Answer the VAK Questionnaire
#### Print the VAK outcomes and determine the dominant learning style

```python
visual_count = 0
auditory_count = 0
kinesthetic_count = 0

# Get user responses to the questions
for i, question in enumerate(questions_lst):
    print(f"{i + 1}. {question}")
#     for j, option in enumerate(unique_responses_by_column[i]):
#         print(f"{chr(65 + j)}) {option}")
    for j, option in enumerate(mcq_options[i]):
        print(f"{chr(65 + j)}) {option}")

    response = input("Your choice (A/B/C): ")
    response = response.lower()
    if (response == 'a'):
        visual_count += 1
    elif (response == 'b'):
        auditory_count += 1
    elif (response == 'c'):
        kinesthetic_count += 1

# Determine the dominant VAK preference
preferences = {
    "Visual": visual_count,
    "Auditory": auditory_count,
    "Kinesthetic": kinesthetic_count
}

dominant_preference = max(preferences, key=preferences.get)

# Print the results
print("\nVAK Learning Style Preferences:")
for style, count in preferences.items():
    print(f"{style}: {count} responses")

print("\nDominant Learning Style:", dominant_preference)
```
