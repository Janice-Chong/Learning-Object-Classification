## Learning Objects Preferrences Based on Dominant VAK

Count all the 1s (Chose 'Very Much' option) for each learning objects based on dominant VAK

```python
import pandas as pd
```

```python
# Load the CSV file containing responses
df = pd.read_csv('Dataset/encoded_response_withQues.csv')
df.head()
```

```python
# Count occurrences of '1' for each learning object per Dominant VAK
learning_objects = df.iloc[:, 3:16]  # Exclude the Dominant_VAK column
learning_objects.head()
```

```python
counts = learning_objects.groupby(df['Dominant_VAK']).sum().reset_index()

counts
```

```python
counts.to_csv('Dataset/eda.csv', index=False) ## for EDA in Tableau
```
